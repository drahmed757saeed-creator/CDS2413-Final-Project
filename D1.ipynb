```python
import pandas as pd

# Load the dataset
data = pd.read_csv("StudentPerformanceFactors.csv")

# Display first 5 rows
data.head()
```

```python
def descriptive_statistics(sample, column):
    stats = {
        "Mean": sample[column].mean(),
        "Median": sample[column].median(),
        "Mode": sample[column].mode()[0],
        "Standard Deviation": sample[column].std(),
        "Minimum": sample[column].min(),
        "Maximum": sample[column].max()
    }
    return stats
```

```python
random_sample = data.sample(n=150, random_state=42)
descriptive_statistics(random_sample, "Previous_Scores")
```

```python
# Systematic sampling: select every 5th record
systematic_sample = data.iloc[::5]

# Apply descriptive statistics on the dependent variable
descriptive_statistics(systematic_sample, "Previous_Scores")
```

```python
import pandas as pd

data = pd.read_csv("StudentPerformanceFactors.csv")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

counts, bins, patches = plt.hist(
    data["Previous_Scores"],
    bins=10,
    color="skyblue",
    edgecolor="black"
)

plt.title("Distribution of Previous Scores", fontsize=16, fontweight="bold")
plt.xlabel("Previous Scores", fontsize=14)
plt.ylabel("Frequency", fontsize=14)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
plt.boxplot(data["Previous_Scores"])
plt.title("Box Plot of Previous Scores")
plt.ylabel("Previous Scores")
plt.show()
```

```python
plt.scatter(data["Hours_Studied"], data["Previous_Scores"])
plt.xlabel("Hours Studied")
plt.ylabel("Previous Scores")
plt.title("Hours Studied vs Previous Scores")
plt.show()
```

```python
import seaborn as sns
numeric_data = data.select_dtypes(include=["int64", "float64"])
sns.heatmap(numeric_data.corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Pearson Correlation
pearson_corr = data["Previous_Scores"].corr(data["Hours_Studied"], method="pearson")

# Spearman Correlation
spearman_corr = data["Previous_Scores"].corr(data["Hours_Studied"], method="spearman")

pearson_corr, spearman_corr
```

```python
from scipy import stats

# One-sample t-test against hypothesized mean = 75
t_statistic, p_value = stats.ttest_1samp(data["Previous_Scores"], 75)

t_statistic, p_value
```
